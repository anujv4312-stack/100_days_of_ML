**3)opening a csv with url**

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

**14) converters**

1.   suppose there is a movie whose full name you dont want instead you want its short form
2.  make a function that converts and pass it in converter function with column name



```python
def rename(name):
  if name =='Why Cheat India':
     return 'WCI'
  else:
    return name
```

```python
rename('Why Cheat India')
```

```python
df = pd.read_csv('/content/movies.csv',converters={'title_x':rename})
df
```
