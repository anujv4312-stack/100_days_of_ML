```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=ab372ae826ab0c495fb01f7c4341871a')
```

```python
df = pd.DataFrame(response.json()['results'])[['genre_ids','title','original_title','vote_average']]
```

```python
df
```

```python
df2 = pd.DataFrame()
```

```python
from numpy import concat

for i in range(1, 552):
    response = requests.get(
        f'https://api.themoviedb.org/3/movie/top_rated?api_key=ab372ae826ab0c495fb01f7c4341871a&page={i}'
    )

    data = response.json()

    if 'results' not in data:
        print(f"Error at page {i}")
        print(data)
        break

    df = pd.DataFrame(data['results'])[
        ['genre_ids', 'title', 'original_title', 'vote_average']
    ]

    df2 = pd.concat([df2, df], ignore_index=True)

print(df2.shape)
```
